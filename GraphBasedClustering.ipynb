```
import numpy as np 
import pandas as pd 
import os
import matplotlib.pyplot as plt
```

```
df=pd.read_csv(base_dir+'ABC.csv')
df.head()
```

# AGE

```
X=df.loc[0:5053,['Prediction_Value','age_x']].values # 'longitude', 'latitude']].values  'gender_x','age_x']].values
```

```
#X[540:560,3]
```

```
#from sklearn.preprocessing import LabelEncoder
#le = LabelEncoder()
#X[:,3]=le.fit_transform(X[:,3])
```

```
#X[540:560,3]
```

```
from sklearn.cluster import SpectralClustering
```

```
nb_clusters=2
```

```
clustering = SpectralClustering(n_clusters=nb_clusters, assign_labels="discretize", random_state=0).fit(X)
y_pred = clustering.labels_
plt.title('Spectral clustering results ')
plt.scatter(X[:, 1], X[:, 0], s=50, c = y_pred);
# centers = clustering.cluster_centers_
# plt.scatter(centers[:, 1], centers[:, 0], c='black', s=200, alpha=0.5);
plt.ylabel("prediction value")
plt.xlabel('age')
plt.show()
```

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y_pred, test_size=0.15, random_state=0)
```

```
X_rem=df.loc[5053:,['Prediction_Value','age_x']].values
```

```
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier(max_depth=5,n_estimators=5, random_state=0)
rf.fit(X_train,y_train)
y_p=rf.predict(X_test)
from sklearn.metrics import accuracy_score
print("acc:Prediction_Value vs age", accuracy_score(y_p,y_test))
preds=rf.predict(X_rem)
print("feature importances: ", rf.feature_importances_)
```

```
plt.title('Random Forest based on Spectral clustering results ')
plt.scatter(X_rem[:, 1], X_rem[:, 0], s=25, c = preds);
plt.ylabel("prediction value")
plt.xlabel('age')
plt.show()
```
